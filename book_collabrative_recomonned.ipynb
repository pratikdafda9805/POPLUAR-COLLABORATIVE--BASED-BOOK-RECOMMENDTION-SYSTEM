Popularity based recommonation model

```python
rating_with_book=rating.merge(book,on='ISBN')
```

```python
count_rating_with_book=rating_with_book.groupby('Book-Title').count()[['Book-Rating']].reset_index()
```

```python
count_rating_with_book.rename(columns={'Book-Rating':'count_rating'},inplace=True)
```

```python
count_rating_with_book=count_rating_with_book[count_rating_with_book['count_rating'] >=250]
```

```python
final_df=count_rating_with_book.merge(rating_with_book,on='Book-Title')
```

```python
top=final_df.groupby('Book-Title')['Book-Rating'].mean().sort_values(ascending=False).reset_index().rename(columns={'Book-Rating':'avg_rating'})
```

```python
top
```

```python
top_50_books=top.merge(books,on=['Book-Title']).sort_values(ascending=False , by='avg_rating').drop_duplicates('Book-Title')[:50]
```
